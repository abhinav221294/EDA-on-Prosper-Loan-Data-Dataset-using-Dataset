```python
condition = (data_pri_load['LoanStatus'] == 'Completed') | (data_pri_load['LoanStatus'] == 'Defaulted') | (data_pri_load['LoanStatus'] == 'Chargedoff')

df = data_pri_load[condition]

df.dropna(inplace=True)

df.reset_index(inplace=True)

df['LoanStatus'].value_counts()
```

```python


def change_to_defaulted (row):    
    if row['LoanStatus'] == 'Chargedoff':        
        return 'Defaulted'    
    else:        
        return row['LoanStatus']    

df['LoanStatus'] = df.apply(change_to_defaulted, axis=1)

df['LoanStatus'].value_counts()
```

```python
data_pri_load['ListingCategory (numeric)'].unique()
```

```python
data_pri_load['ListingCategory (numeric)'].value_counts()
```

```python
"""
Step 1: Create a  dictionary called categories = 1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto',
7: 'Other'

Step2: Create a user define function  using categories and `ListingCategory (numeric)` column.

Hint: df['ListingCategory (numeric)'] = df.apply(user define function , axis=1)

    Sample output : 


    Debt Consolidation    106

    Other                  65

    Business               25

    Home Improvement       22

    Auto                    9


"""

dict1 = {1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto',7: 'Other'}

ListingCategory = df['ListingCategory (numeric)']

def listingcategory(ListingCategory):
    
    if ListingCategory == 1:
        return dict1.get(1)
    elif ListingCategory == 2 :
        return dict1.get(2)
    elif ListingCategory == 3:
        return dict1.get(3)
    elif ListingCategory == 6:
        return dict1.get(6)
    else:
        return dict1.get(7)
    
    
df['ListingCategory (numeric)'] = df['ListingCategory (numeric)'].apply(listingcategory)    
```

```python
df['ListingCategory (numeric)'].value_counts()
```

```python
# C. Display the box plot for ProsperRating (Alpha) vs LoanOriginalAmount and hue = Loan Status[Completed,Defaulted].
# Write your observations.

plt.rcParams["figure.figsize"] = (20,10)
sns.boxplot(data=df,x='ProsperRating (Alpha)',y='LoanOriginalAmount',hue='LoanStatus')
plt.show()
```
